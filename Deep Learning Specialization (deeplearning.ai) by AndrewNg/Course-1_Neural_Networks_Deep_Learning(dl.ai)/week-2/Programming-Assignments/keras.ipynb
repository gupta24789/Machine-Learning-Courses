```python
import numpy as np
import pandas as pd
import h5py
import matplotlib.pyplot
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
## Load the data first
train_org  = h5py.File('./datasets/train_catvnoncat.h5')
test_org = h5py.File('./datasets/test_catvnoncat.h5')
```

```python
## set the keys available in train and test data
print("Train data keys :",[key for key in train_org.keys()])
print("Test data keys :",[key for key in test_org.keys()])
```

```python
train_org_x = np.array(train_org['train_set_x'])
train_org_y = np.array(train_org['train_set_y'])

test_org_x = np.array(test_org['test_set_x'])
test_org_y = np.array(test_org['test_set_y'])

list_classes = np.array(train_org['list_classes'])
```

```python
## Check the shape of train and test data
print("Train X shape :",train_org_x.shape)
print("Train Y shape :",train_org_y.shape)
print("Test X shape :",test_org_x.shape)
print("Test Y shape :",test_org_y.shape)
print("Classes :", list_classes)
```

```python
## Create feature vector
train_x  = train_org_x.reshape(train_org_x.shape[0],-1)
train_y  = train_org_y.reshape(train_org_y.shape[0],1)

test_x  = test_org_x.reshape(test_org_x.shape[0],-1)
test_y  = test_org_y.reshape(test_org_y.shape[0],1)
```

```python
## Again check the shape 
print("Train X shape :",train_x.shape)
print("Train Y shape :",train_y.shape)

print("Test X shape :",test_x.shape)
print("Test Y shape :",test_y.shape)
```

```python
## normalize the input vectors 
train_x = train_x/255
test_x = test_x/255
```

```python
from keras.models import  Sequential
from keras.layers import Dense, Activation, Dropout
from keras.optimizers import SGD, Adam
import keras
```

## Single layer NN 

```python
import tensorflow as tf
```

```python
tf.set_random_seed(1)
np.random.seed(1)
```

```python
model = Sequential()
model.add(Dense(units=20, activation= 'relu', input_shape = (train_x.shape[1],)))
model.add(Dense(units=10, activation= 'relu'))
model.add(Dense(units=5, activation= 'relu'))
model.add(Dense(units=1,activation='sigmoid'))
```

```python
train_accuracy = []
test_accuracy = []

for lr in [.005,.0075]:
    sgd = SGD(lr=lr)
    model.compile(optimizer= sgd, loss= keras.losses.binary_crossentropy, metrics=['accuracy'])
    model.fit(x = train_x, y= train_y,verbose=1, batch_size=32, epochs=2500)

    train_accuracy.append(model.evaluate(x=train_x,y=train_y, batch_size=32)[1])
    test_accuracy.append(model.evaluate(x=test_x,y=test_y, batch_size=32)[1])
    
```

```python
print(train_accuracy)
print(test_accuracy)
```
